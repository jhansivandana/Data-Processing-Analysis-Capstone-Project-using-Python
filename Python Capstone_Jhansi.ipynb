```python
#Task 2 : Fill in the missing values in Cost column by running average using “For” loop for this task.
df = pd.read_csv('project_dataframe.csv')    # This code will read the 'project_dataframe.csv' file and enables us to access the data
```

```python
df['Cost'] = pd.to_numeric(df['Cost'], errors='coerce')    # While building dataframe although they are mapped using np.nan for missing values, it could not read the value. Hence the parameter errors='coerce' converts these values to NaN (Not a Number) which is a special marker for missing values.
```

```python
running_sum = 0               # Initialising data for "for" loop  
count = 0                     # Initialising data for "for" loop 
costs = df['Cost'].tolist()   # The code converts the 'Cost' column in the DataFrame to a Python list for easier iteration
for i in range(len(costs)):   # Range starts from '0' to '13' which is the total lenght of the dataframe
    if pd.isna(costs[i]):     # pd.isna() check if the value is NaN or not for each i
        if count > 0:         #if count > 0 then it enters if statement, if not it enters else statement
            costs[i] = running_sum / count    # Computing average
        else:
            costs[i] = 0  # or another logic if all previous values are missing
    else:
        running_sum += costs[i]    # Stepping to next iteration
        count += 1                 # Stepping to next iteration
df['Cost'] = costs                 # Assigning costs to df(Cost)
df['Cost'] = df['Cost'].round(2)   # Rounding of the decimal place to 2
```

```python
df.to_csv('project_dataframe_updated.csv', index=False)   # Saving updated dataframe 
```

```python
#Task 4 : Join all three dataframes in one single dataframe with the name “Final”
# Reading all the dataframes to perform the required task
df = pd.read_csv('project_dataframe_updated.csv')
edf = pd.read_csv('employee_dataframe_updated.csv')
sdf = pd.read_csv('seniority_dataframe.csv')
```

```python
df_edf_merged = pd.merge(df, edf, on='ID', how='left')     # joining two dataframes with a primary key 'ID'
```

```python
Final = pd.merge(df_edf_merged,sdf, on='ID', how='left')   # joining the merged dataframes with sdf dataframe through a primary key 'ID'
```

```python
Final.to_csv('Final.csv', index=False)                     # saving the merged dataframe as 'Final.csv' file
```

```python
#Task 5 : Give a 5% bonus (new column) concerning project cost only to employees who have finished the projects.
# Creating new column called bonus and giving 5% bonus to employees with finished project status
Final['Bonus'] = Final.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)
```

```python
#Task 8 : Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.
count = 0                          # Starting the iteration count from '0'
for i in range(len(df)):           # Range is considered from length of the dataframe (0-13)
    if df.loc[i, 'Age'] > 29:      # checking for 'Age > 29'. If the condition is 'True' then it moves to the next statement.
        df.loc[i, 'Designation Level'] -= 1     # Here, the designation level is promoted by 1 level. 
    else:                          # if the 'IF' statement returns 'False', then else statement is executed.
        count += 1                 # And iteration count is incremented by 1
```
